### DATA VISUALIZATION

```python
plt.figure(figsize=(10,5))
c= retail.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
sns.regplot(x="Profit",y="Discount",data=retail)
plt.ylim(0,)
```

**We can see a negative correlation between Discount and Profit**

```python
sns.regplot(x="Profit",y="Sales",data=retail)
plt.ylim(0,)
```

**There is a Positive correlation between Sales and Profit**

```python
plt.figure(figsize=(12,6))
plt.bar('Sub-Category','Category', data=retail)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Catgory')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,12))
retail['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%').set_title('SUB CATEGORIES')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.countplot(x=retail['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
ps = retail.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(figsize=(15,10))
plt.title('Profit/Loss and Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

- Highest Sales is for Binders,Papers and Furnishings 
- Lowest is for Supplies,Machines and Copiers
- Phones,Chairs,Storage has highest Sales in sub category
- Tables,Bookcases,Supllies are the sub categories that has least Profit


```python
plt.figure(figsize=(10,15))
sns.countplot(x=retail['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].set_title('Discount vs Profit', fontsize=15)
ax[1].set_title('Quantity vs Profit', fontsize=15)
sns.lineplot(retail.Discount, retail.Profit, color='green', label='Profit Change', ax=ax[0])
sns.lineplot(retail.Quantity, retail.Profit, color='red', label='Profit Change', ax=ax[1])
plt.show()
```

```python
sns.countplot(x=retail['Segment'])
```

As we can see in this graph highest Segment is Consumer Segment following with Corporate & Home Office

```python
sns.countplot(x='Category',data=retail).set_title('CATEGORY')
```

```python
retail.groupby('Category')['Sales','Profit'].sum().plot.bar()
plt.title('Sale & profit in different categories')
plt.legend(loc=1)
plt.show()
```

- Technology & Furniture has more sales than Office Supllies
- In case of Profit Furniture has least value

```python
sns.countplot(x=retail['Region'])
```

Highest Sales Occurs at West & East Regions

```python
segment2=retail[['Category','Region','Profit']]
reg_analysis=segment2.groupby(['Category','Region'],as_index=False).mean()
```

```python
pivot=reg_analysis.pivot(index='Region',columns='Category')
pivot
```

```python
ax = sns.heatmap(pivot, annot=True,linewidth=.05)
```

- Technology Category has the highest profit at East Region 
- Furniture has the least profit at Central Region

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(retail["Sales"] , retail["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
data = retail[(retail['Sub-Category'] == 'Bookcases') | (retail['Sub-Category'] == 'Tables')].sort_values('Profit')
plt.figure(figsize=(15,5))
plt.title('Overall Profit - Tables & Bookcases', fontsize=15)
sns.barplot(data['State'], data.Profit)
plt.xticks(rotation=75)
plt.show()
```

```python
ps = retail.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(figsize=(15,10))
plt.title('SALES & PROFIT ACROSS STATES')
plt.xlabel('States')
plt.ylabel('Sales/Profit')
plt.show()
```

Highest Sales is at California and New York Sates while Lowest Profit is at Texas and Ohio States

```python
retail.groupby('Region')['Sales','Profit'].sum().plot.bar()
plt.title('Sale & profit in each region')
plt.legend()
plt.show()
```

East and west are better performing regions in sales and profit
